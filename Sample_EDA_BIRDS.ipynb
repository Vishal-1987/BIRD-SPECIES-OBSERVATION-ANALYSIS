## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Analyze the Date and Year columns to detect patterns in bird sightings across different seasons or years.

````python
# changing date time format
df['Date_x']=pd.to_datetime(df['Date'],format='%y-%m-%d')

# extracting year and month 
df['year']=df['Date_x'].dt.year
df['month']=df['Date_x'].dt.month
fig, ax = plt.subplots(1, 2,figsize=(12, 5))

# birds sighting by year
year_sighting=df['year'].value_counts()
sns.lineplot(ax=ax[0],x=year_sighting.index,y=year_sighting.values,marker='o')
ax[0].set_title('YEAR WISE SIGHTINGS',fontsize=14)
ax[0].set_xlabel('YEAR')
ax[0].set_ylabel('TOTAL SIGHTINGS')
ax[0].grid(True)

# birds sighting by month
month_sighting=df['month'].value_counts()
plot=sns.barplot(ax=ax[1],x=month_sighting.index,y=month_sighting.values,width=0.2,color='green')
for label in plot.patches:
    height=label.get_height()
    ax[1].text(label.get_x()+label.get_width()/2,height,str(int(height)),ha='center',va='bottom')
ax[1].set_title('MONTH WISE SIGHTINGS',fontsize=14)
ax[1].set_xlabel('MONTH')
ax[1].set_xlabel('TOTAL SIGHTINGS')
plt.show()
````

#####  What is/are the insight(s) found from the chart?

#### The chart shows that bird sightings vary noticeably by month, with peaks in certain months indicating seasonal activity or migration periods, and lower counts in off-peak months.

#### Chart - 2

#### Study the Start_Time and End_Time to determine if specific time windows correlate with higher bird activity

````python
# changing the time format
df['Start_Time']=pd.to_datetime(df['Start_Time'],format='%H:%M:%S')
df['start_hour']=df['Start_Time'].dt.hour
df['End_Time']=pd.to_datetime(df['End_Time'],format='%H:%M:%S')
df['end_hour']=df['End_Time'].dt.hour

# creating side by side plot
fig,ax=plt.subplots(1,2,figsize=(13,6))

# counting bird activity by start time
start_activity=df['start_hour'].value_counts()
sns.lineplot(ax=ax[0],x=start_activity.index,y=start_activity.values)
x=start_activity.index
y=start_activity.values
for i in range(len(x)):
    ax[0].text(x[i],y[i]-1,int(y[i]),ha='center', va='bottom')
ax[0].set_title('BIRD ACTIVITY BY START TIME')
ax[0].set_xlabel('STARTING_HOUR')
ax[0].set_ylabel('ACTIVITY RATE')
ax[0].set_facecolor('#eaffea')


# counting bird activity by end time
end_activity=df['end_hour'].value_counts()
sns.lineplot(ax=ax[1],x=end_activity.index,y=end_activity.values)
x=end_activity.index
y=end_activity.values
for i in range(len(x)):
    ax[1].text(x[i],y[i]-1,int(y[i]),ha='center', va='bottom')
ax[1].set_title('BIRD ACTIVITY BY END TIME')
ax[1].set_xlabel('ENDING_HOUR')
ax[1].set_ylabel('ACTIVITY RATE')
ax[1].set_facecolor('#ffeaea')
plt.show()
````

#### What is/are the insight(s) found from the chart?

##### Morning Peak – Highest bird activity starts at 6 AM with 45 observations, followed by 7 AM with 38 observations.
##### Evening End – Most activities end around 6 PM (40 observations) and 5 PM (35 observations).
##### Active Window – Majority of bird activities occur between 5 AM and 7 PM, covering nearly 85% of total records.

#### Chart - 11

#### Evaluate the Visit column to see how repeated visits affect species count or diversity.

````python
total_visit=df.groupby('Visit')['Common_Name'].nunique().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,6))
sns.set_style('dark')
plot=sns.barplot(data=total_visit,x='Visit',y='Common_Name',hue='Visit',palette='viridis',width=0.5)
for label in plot.patches:
    height=label.get_height()
    if height==0:
        continue
    plt.text(label.get_x()+label.get_width()/2,height,str(int(height)),ha='center',va='bottom')
plt.title('VISIT WISE SPECIES COUNT')
plt.xlabel('VISIT')
plt.ylabel('BIRDS COUNT')
plt.show()
````

##### 2. What is/are the insight(s) found from the chart?

##### Visit 1 recorded the highest species diversity 113 species.
##### Visit 2 recorded 94 species, showing a small/mild drop from Visit 1.
##### Later visits (Visit 3 have 68 species observed, suggesting reduced diversity with repeated visits

#### Chart - 12

#### Use the PIF_Watchlist_Status and Regional_Stewardship_Status to identify trends in species that are at risk or require conservation focus.

````python
# Count combinations of PIF_Watchlist_Status and Regional_Stewardship_Status
status_counts = df.groupby(['PIF_Watchlist_Status', 'Regional_Stewardship_Status']).size().reset_index(name='Count')

# Plot
plt.figure(figsize=(12,6))
sns.barplot(data=status_counts, 
            x='PIF_Watchlist_Status', 
            y='Count', 
            hue='Regional_Stewardship_Status', 
            palette='Set2')

plt.title('Species at Risk & Conservation Focus')
plt.xlabel('PIF Watchlist Status')
plt.ylabel('Species Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Regional Stewardship Status')
plt.tight_layout()
plt.show()
````
